##### b. Understanding how many new customers acquired every month


```python
Orders['purchase_month'] = Orders.order_purchase_timestamp.dt.strftime('%Y-%m')
```

```python
cust_orders = Customers.merge(Orders, how = 'left', on = 'customer_id')
```

```python
cust_orders.head()
```

```python
cust_orders_new = cust_orders.drop_duplicates(subset = 'customer_unique_id', keep = 'first')
```

```python
new_customer_acquired_every_month = cust_orders_new.groupby('purchase_month')['customer_unique_id'].count().reset_index()
```

```python
new_customer_acquired_every_month = new_customer_acquired_every_month.rename(columns = {'purchase_month' : 'Month', 
                                                    'customer_unique_id' : 'No. of Customers acquired'}).set_index('Month')
```

```python
new_customer_acquired_every_month
```

##### c. Understand the retention of customers on month on month basis


```python
total_customers_every_month = cust_orders.groupby('purchase_month')['customer_unique_id'].count().reset_index()
```

```python
total_customers_every_month = total_customers_every_month.rename(columns = {'purchase_month' : 'Month', 
                                              'customer_unique_id' : 'No. of Customers'}).set_index('Month')
```

```python
total_customers_every_month
```

```python
retained_cust = total_customers_every_month.merge(new_customer_acquired_every_month, how = 'left', on = 'Month')
```

```python
retained_cust.fillna(0, inplace = True)
```

```python
retained_cust['Retained Customers'] = retained_cust['No. of Customers'] - retained_cust['No. of Customers acquired']
```

```python
retained_cust.drop(columns = (['No. of Customers', 'No. of Customers acquired']), inplace = True)
```

```python
retained_cust
```

### 3. Cross-Selling (Which products are selling together)

```python
from itertools import combinations
from collections import Counter
```

```python
prod_selling_together = cust_order_details.dropna()
prod_selling_together = prod_selling_together[prod_selling_together['order_id'].duplicated(keep = False)]
prod_selling_together['products'] = prod_selling_together.groupby('order_id')['product_id'].transform(lambda x:','.join(x))
prod_selling_together = prod_selling_together[['order_id', 'products']].drop_duplicates()
prod_selling_together.head()
```

```python
count = Counter()
for i in prod_selling_together['products']:
    row_list = i.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```
